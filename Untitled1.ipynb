```python
# Step 0: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#  Step 1: Load the EV charging dataset
data_file = "ev_charging_patterns.csv"
df = pd.read_csv(data_file)

# Step 2: Quick sanity check on the data
print("Shape of data:", df.shape)
print("\nColumn types:\n", df.dtypes)
print("\nNull values per column:\n", df.isna().sum())
print("\nFirst few rows just to see what it looks like:\n", df.head(7))  # I like 7 instead of 5

#  Step 3: Descriptive Stats
print("\nSummary (numeric only):\n")
print(df.describe(include=[np.number]))

# Step 4: Data Cleaning
df = df.drop_duplicates()

# filling missing energy consumption if that column exists
if "Energy (kWh)" in df.columns:
    avg_energy = df["Energy (kWh)"].mean()
    df["Energy (kWh)"] = df["Energy (kWh)"].fillna(avg_energy)

# Step 5: Feature Engineering
if "Start Time" in df.columns:
    df["Start Time"] = pd.to_datetime(df["Start Time"], errors="coerce")  # coerce just in case of bad formats
    df["Hour"] = df["Start Time"].dt.hour
    df["Weekday"] = df["Start Time"].dt.weekday
# Step 6: Visual Exploration

# 1. Distribution of charging demand
if "Energy (kWh)" in df.columns:
    plt.figure(figsize=(9,5))
    sns.histplot(df["Energy (kWh)"], bins=25, kde=True, color="skyblue")
    plt.title("How much energy is consumed per charging session?")
    plt.xlabel("Energy (kWh)")
    plt.ylabel("Number of sessions")
    plt.grid(alpha=0.3)
    plt.show()

# 2. Average demand by hour
if "Hour" in df.columns and "Energy (kWh)" in df.columns:
    hourly_means = df.groupby("Hour")["Energy (kWh)"].mean()
    plt.figure(figsize=(10,5))
    hourly_means.plot(kind="bar", color="orange", edgecolor="black")
    plt.title("Avg Charging Energy by Hour of Day")
    plt.xlabel("Hour")
    plt.ylabel("Avg Energy (kWh)")
    plt.axhline(hourly_means.max(), color="red", linestyle="--", alpha=0.6)
    plt.show()

# 3. Correlation heatmap of numeric columns
plt.figure(figsize=(8,6))
# check if categorical encoding is required
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Features")
plt.show()

```
